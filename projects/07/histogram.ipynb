```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("mask1.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
hist = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
hist = cv2.calcHist([img],[0],None,[256],[0,256])

plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
```
